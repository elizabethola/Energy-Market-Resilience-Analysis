### Data Cleaning

```python
#grouping all datasets as dataset
datasets = {
    "historical_energy_data": historical_energy_data,
    "infrastructure_data": infrastructure_data,
    "market_data": market_data,
    "regulatory_data": regulatory_data
}

#Handling Duplicate Rows
#We'll start by checking each dataset for duplicate rows. If duplicates are found, we'll remove them.

# Removing duplicate rows from each dataset and storing the number of duplicates removed
duplicates_removed = {}

for name, data in datasets.items():
    before_removal = len(data)
    data.drop_duplicates(inplace=True)
    after_removal = len(data)
    duplicates_removed[name] = before_removal - after_removal

duplicates_removed
```

No duplicate rows were found in any of the datasets.

```python
#Investigating Categorical Columns
#Next, we'll take a closer look at the categorical columns in each dataset to identify any inconsistencies or anomalies. Specifically, we'll examine:
 #Location/Region and Energy Source from historical_energy_data
 #Infrastructure Status, Maintenance Activities, and Technology Limitations from infrastructure_data
 #Competitor Data and Market Trends from market_data
 #Regulatory Changes and Compliance Status from regulatory_data
 #Let's view the unique values and their counts for each of these categorical columns.
# Extracting unique values and their counts for categorical columns
categorical_columns = {
    "historical_energy_data": ["Location/Region", "Energy Source"],
    "infrastructure_data": ["Infrastructure Status", "Maintenance Activities", "Technology Limitations"],
    "market_data": ["Competitor Data", "Market Trends"],
    "regulatory_data": ["Regulatory Changes", "Compliance Status"]
}

categorical_values_counts = {}
for dataset, columns in categorical_columns.items():
    for column in columns:
        categorical_values_counts[(dataset, column)] = datasets[dataset][column].value_counts()
```

```python
#To get a visual understanding of the data's categorical columns
# Calculate the number of rows and columns for the subplots
num_rows = (len(categorical_values_counts) + 1) // 2
num_cols = 2

# Create a grid of subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 12))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Iterate through categorical columns and plot them in subplots
for i, ((dataset, column), counts) in enumerate(categorical_values_counts.items()):
    ax = axes[i]
    sns.barplot(x=counts.values, y=counts.index, ax=ax)
    ax.set_title(f"Value Counts for {dataset} - {column}")
    ax.set_xlabel("Count")
    ax.set_ylabel(column)

# Hide any remaining empty subplots
for i in range(len(categorical_values_counts), num_rows * num_cols):
    fig.delaxes(axes[i])

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()
```

From the above details, the categorical values seem consistent and don't show any apparent anomalies.

### Exploratory Data Analysis

```python
#Energy Demand, Production, and Consumption Over Time 
#We'll start by examining the trends in energy demand, production, and consumption from the historical_energy_data dataset. 
#This will give us insights into fluctuations and potential vulnerabilities in the energy market.import pandas as pd


# Extract the month and year from the 'Date/Time' column
historical_energy_data['Year'] = historical_energy_data['Date/Time'].dt.year
historical_energy_data['Month'] = historical_energy_data['Date/Time'].dt.month

# Create a new column 'Year-Month' for easy plotting
historical_energy_data['Year-Month'] = historical_energy_data['Date/Time'].dt.to_period('M')

# Aggregating data on a monthly basis
monthly_data = historical_energy_data.groupby('Year-Month').mean()

# Setting up the figure
plt.figure(figsize=(15, 8))

# Plotting Energy Production, Consumption, and Demand over time (monthly)
sns.lineplot(data=monthly_data, x=monthly_data.index.astype(str), y='Energy Production (kWh)', label='Energy Production', color="blue", linestyle='-', linewidth=1.5, ci= None)
sns.lineplot(data=monthly_data, x=monthly_data.index.astype(str), y='Energy Consumption (kWh)', label='Energy Consumption', color="red", linestyle='-', linewidth=1.5, ci= None)
sns.lineplot(data=monthly_data, x=monthly_data.index.astype(str), y='Energy Demand', label='Energy Demand', color="green", linestyle='-', linewidth=1.5, ci= None)

# Setting the title and labels
plt.title('Monthly Aggregate of Energy Production, Consumption, and Demand Over Time')
plt.xlabel('Date')
plt.ylabel('kWh')
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

labels = monthly_data.index.astype(str).tolist()
n = 6  
plt.xticks(labels[::n], rotation=360) 
plt.tight_layout()

# Adjust the layout
plt.tight_layout()
plt.subplots_adjust(hspace=0.5)

plt.show()
```

```python
# Zooming in to have a closer look in years.
# Setting 'Date/Time' as the index of the DataFrame
# Extract the year and month from the 'Date/Time' column
historical_energy_data['Year'] = historical_energy_data['Date/Time'].dt.year
historical_energy_data['Month'] = historical_energy_data['Date/Time'].dt.month

# Get unique years in the DataFrame
unique_years = historical_energy_data['Year'].unique()

# Setting up the figure
plt.figure(figsize=(10, 10))

for index, year in enumerate(unique_years, 1):
    # Selecting data for the specific year
    yearly_data = historical_energy_data[historical_energy_data['Year'] == year]
    
    # Creating a subplot for each year
    plt.subplot(len(unique_years), 1, index)
    
    
# Plotting Energy Production, Consumption, and Demand over time (monthly) for the specific year
    sns.lineplot(data=yearly_data, x='Month', y='Energy Production (kWh)', label='Energy Production', color="blue", linestyle='-', linewidth=1.5, ci= None)
    sns.lineplot(data=yearly_data, x='Month', y='Energy Consumption (kWh)', label='Energy Consumption', color="red", linestyle='-', linewidth=1.5, ci= None)
    sns.lineplot(data=yearly_data, x='Month', y='Energy Demand', label='Energy Demand', color="green", linestyle='-', linewidth=1.5, ci= None)
    
    # Setting the title and labels for each subplot
    plt.title(f'Monthly Aggregate of Energy Production, Consumption, and Demand in {year}')
    plt.xlabel('Month')
    plt.ylabel('kWh')
    plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    
    # Adjust the layout
plt.tight_layout()
plt.subplots_adjust(hspace=0.5)

plt.show()
```

From the graph, we observe:

```python
#Market Price and Energy Price Over Time
#To understand the pricing dynamics and its relation to market trends,
#let's visualize the Market Price from the market_data dataset alongside the Energy Price from the historical_energy_data dataset over time.
#Convert Market data date/time column to datetime
market_data['Date/Time'] = pd.to_datetime(market_data['Date/Time'])


# Merging historical_energy_data and market_data on Date/Time for combined analysis
merged_data = pd.merge(historical_energy_data, market_data, on="Date/Time", how="inner")

# Setting 'Date/Time' as the index of the DataFrame for aggregation
merged_data['Year'] = merged_data['Date/Time'].dt.year
merged_data['Month'] = merged_data['Date/Time'].dt.month

# Aggregating data on a monthly basis
monthly_merged_data = merged_data.groupby(['Year', 'Month']).mean()

# Create a new 'Year-Month' column for plotting
monthly_merged_data['Year-Month'] = monthly_merged_data.index.get_level_values(0).astype(str) + '-' + monthly_merged_data.index.get_level_values(1).astype(str)

# Setting up the figure and axes
plt.figure(figsize=(20, 10))

# Plotting Market Price and Energy Price over time (monthly) for all years combined
sns.lineplot(data=monthly_merged_data, x='Year-Month', y='Market Price', label='Market Price', color="blue", linestyle='-', linewidth=1.5, ci= None)
sns.lineplot(data=monthly_merged_data, x='Year-Month', y='Energy Price', label='Energy Price', color="yellow", linestyle='-', linewidth=1.5, ci= None)

# Setting the title and labels
plt.title('Monthly Aggregate of Market Price and Energy Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.grid(True, which='both', linestyle='-', linewidth=0.5)
labels = monthly_merged_data['Year-Month'].tolist()
n = 4  
plt.xticks(labels[::n], rotation=360)
plt.tight_layout()

plt.show()
```

```python
# Convert 'Date/Time' columns in both dataframes to datetime
historical_energy_data['Date/Time'] = pd.to_datetime(historical_energy_data['Date/Time'])
market_data['Date/Time'] = pd.to_datetime(market_data['Date/Time'])

# Merge the two dataframes
merged_data = pd.merge(historical_energy_data, market_data, on="Date/Time", how="inner")

# Extract the year and month from the 'Date/Time' column
merged_data['Year'] = merged_data['Date/Time'].dt.year
merged_data['Month'] = merged_data['Date/Time'].dt.month

# Get unique years in the DataFrame
unique_years = merged_data['Year'].unique()

# Setting up the figure with subplots for each year
fig, axes = plt.subplots(len(unique_years), 1, figsize=(20, 6 * len(unique_years)))

for index, year in enumerate(unique_years):
    yearly_data = merged_data[merged_data['Year'] == year]
    grouped_yearly_data = yearly_data.groupby(['Month']).mean()
    
    grouped_yearly_data['Year-Month'] = grouped_yearly_data.index.astype(str) + '-' + year.astype(str)
    
    sns.lineplot(data=grouped_yearly_data, x='Month', y='Market Price', ax=axes[index], label='Market Price', color="blue", linestyle='-', linewidth=1.5, ci=None)
    sns.lineplot(data=grouped_yearly_data, x='Month', y='Energy Price', ax=axes[index], label='Energy Price', color="yellow", linestyle='-', linewidth=1.5, ci=None)
    
    # Setting the title for each subplot
    axes[index].set_title(f'Market Price and Energy Price in {year}')
    axes[index].set_xlabel('Month')
    axes[index].set_ylabel('Price ($)')
    axes[index].legend()
    axes[index].grid(True, which='both', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```

The graph showcases the trends in Market Price and Energy Price over time:

```python
#Correlation between poor infrstructure status and high tech limitation and the energy production
# Convert 'Infrastructure Status' and 'Technology Limitations' to binary columns
infrastructure_data['Poor_Infrastructure'] = infrastructure_data['Infrastructure Status'].apply(lambda x: 1 if x == 'Poor' else 0)
infrastructure_data['High_Tech_Limitations'] = infrastructure_data['Technology Limitations'].apply(lambda x: 1 if x == 'High' else 0)

# Create a combined column 
infrastructure_data['Combined'] = infrastructure_data['Poor_Infrastructure'] * infrastructure_data['High_Tech_Limitations']

# Merge infrastructure_data with historical_energy_data on Date/Time
merged_data = pd.merge(historical_energy_data, infrastructure_data, on="Date/Time", how="inner")

# Calculate correlation coefficient
correlation_coefficient = merged_data['Energy Production (kWh)'].corr(merged_data['Combined'])

correlation_coefficient
```

The correlation coefficient of −0.0159 indicates a very weak negative linear relationship between "Energy Production" and instances where both the "Infrastructure Status" is "Poor" and "Technology Limitations" is "High".

In practical terms, this means that there is a slight tendency for energy production to decrease when both the infrastructure status is poor and the technology limitations are high.

```python
#Impact of Regulatory Changes and Operational Costs on Revenue
#We'll visualize the relationship between regulatory changes and the associated compliance costs, as well as the operational costs to the Revenue generated from the company.
# Revenue = Energy Consumption* Energy price
#This will provide insights into the financial implications of regulatory changes cost and operational cost for Energix Enterprise.

# Calculate Revenue
historical_energy_data['Revenue'] = historical_energy_data['Energy Price'] * historical_energy_data['Energy Consumption (kWh)']

# Merge historical_energy_data and regulatory_data on Date/Time
merged_regulatory_data = pd.merge(historical_energy_data, regulatory_data, on="Date/Time", how="inner")

# Extract year and month from 'Date/Time'
merged_regulatory_data['Year'] = merged_regulatory_data['Date/Time'].dt.year
merged_regulatory_data['Month'] = merged_regulatory_data['Date/Time'].dt.month

# Group by year and month and compute mean for each month
monthly_aggregated_data = merged_regulatory_data.groupby(['Year', 'Month']).mean().reset_index()

# Create a 'Year-Month' column for plotting
monthly_aggregated_data['Year-Month'] = monthly_aggregated_data['Year'].astype(str) + '-' + monthly_aggregated_data['Month'].astype(str)

# Plotting
plt.figure(figsize=(20, 8))

# Stacked area chart for Operational Costs and Compliance Costs
plt.fill_between(monthly_aggregated_data['Year-Month'], monthly_aggregated_data['Operational Costs'], color='blue', label='Operational Costs', alpha=0.5)
plt.fill_between(monthly_aggregated_data['Year-Month'], monthly_aggregated_data['Operational Costs'], monthly_aggregated_data['Operational Costs'] + monthly_aggregated_data['Compliance Costs'], color='green', label='Compliance Costs', alpha=0.5)

# Line plot for Revenue
sns.lineplot(data=monthly_aggregated_data, x='Year-Month', y='Revenue', label='Revenue', color='red', linewidth=2)

# Setting the title and labels
plt.title('Impact of Regulatory Changes on Operational, Revenue, and Compliance Costs (Monthly Aggregated)')
plt.xlabel('Date')
plt.ylabel('Amount ($)')
plt.legend(loc='upper left')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
labels = monthly_aggregated_data['Year-Month'].tolist()
n = 4  
plt.xticks(labels[::n], rotation=360)

# Show the plot
plt.show()
```
